### Load CO2 data

The CO2 data is retrieved from the GitHub page ["of our World in data"](https://github.com/owid/co2-data). A lot of interesting variables, but also a ton of missing data. Will be interesting to check out. Data is also only available on a yearly instead of a monthly basis. Also Something we should have a look at.

```python
CO2_df = pd.read_csv(filepath_or_buffer ="D:\Data\Dropbox\LifeAfter\Datascientest\Climate\Data\CO2_data\owid-co2-data.csv")
CO2_df.info()
```

Looking at the variables main variables of interest, which in my eyes are: cement_co2, co2, coal_co2, consumption_co2, primary_energy_consumption, flaring_co2, gas_co2, land_use_change_co2, methane, nitrous_oxide, oil_co2, other_industry_co2, trade_co2, total_ghg. First we plot the original distribution:

```python
var_int = ["cement_co2", "co2", "coal_co2", "consumption_co2", "primary_energy_consumption", "flaring_co2",
           "gas_co2", "land_use_change_co2", "methane", "nitrous_oxide", "oil_co2", "other_industry_co2", 
           "trade_co2", "total_ghg"]

warnings.filterwarnings('ignore')
# plot the distribution of all of these variablesabs

fig, axes = plt.subplots(5, 3, figsize=(18, 18))
row_c = 0
col_c = 0
for plot in range(len(var_int)):
    if col_c ==3:
        col_c = 0
        row_c+=1
    sns.histplot(data = np.log(CO2_df[var_int[plot]]), kde=True, stat='density', ax = axes[row_c, col_c])
    plt.xlabel(var_int[plot] + "log-transformed")
    col_c +=1
```
